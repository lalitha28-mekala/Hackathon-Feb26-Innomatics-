```python
import pandas as pd
import numpy as np
```

```python
# SQL file import in to colab and read as df
import sqlite3
conn=sqlite3.connect("restaurants.db")
cursor=conn.cursor()

# read the sql scripts
with open("restaurants.sql","r") as f:
  sql_scripts=f.read()
  cursor.executescript(sql_scripts)
  conn.commit()

df3=pd.read_sql("select * from restaurants",conn)
```

```python
df3.head()
```

```python
df1=pd.read_csv("/content/orders.csv")
df2=pd.read_json("/content/users.json")
```

```python
df1.head()
```

```python
df2
```

```python
x=pd.merge(df1,df2,on="user_id",how="left")
y=pd.merge(x,df3, on="restaurant_id",how="left")
```

```python
y
```

```python
y.info()
```

```python
y["order_date"]=pd.to_datetime(y["order_date"],errors="coerce",dayfirst=True)
```

```python
y.info()
```

```python
y.head()
```

```python
# q1 Which city has the highest total revenue (total_amount) from Gold members?
df_gold=y[y["membership"]=="Gold"]
df_gold.groupby("city")["total_amount"].sum().sort_values(ascending=False)
```

```python
# q2 Which cuisine has the highest average order value across all orders?
y.groupby("cuisine")["total_amount"].mean().sort_values(ascending=False)
```

```python
# Q3 How many distinct users placed orders worth more than ₹1000 in total (sum of all their orders)?
total_order_cut=y.groupby("user_id")["total_amount"].sum().reset_index()
total_order_cut[total_order_cut["total_amount"]>1000]
```

```python
# Q4 Which restaurant rating range generated the highest total revenue?
bins = [3.0, 3.5, 4.0, 4.5, 5.0]
labels = ['3.0–3.5','3.6–4.0','4.1–4.5','4.6–5.0']

y['rating_range'] = pd.cut(y['rating'], bins=bins, labels=labels)

y.groupby('rating_range')['total_amount'].sum().sort_values(ascending=False)
```

```python
# Q5 Among Gold members, which city has the highest average order value?
df_gold.groupby("city")["total_amount"].mean().sort_values(ascending=False)
```

```python
y.head()
```

```python
# Q6 Which cuisine has the lowest number of distinct restaurants but still contributes significant revenue?
cuisine_summ=y.groupby(["cuisine","restaurant_id"]).agg(amt=("total_amount","sum"),count=("restaurant_id","nunique")).reset_index()
```

```python
cuisine_summ.sort_values(
    by=["count","amt"],
    ascending=[True, False])
```

```python
cuisine_summ["count"].unique()
```

```python
# Q7 What percentage of total orders were placed by Gold members? (Rounded to nearest integer)

total=(df_gold["total_amount"].sum())/(y["total_amount"].sum())*100
total
```

```python
# Q8 Which restaurant has the highest average order value but less than 20 total orders?
res_info=y.groupby("restaurant_name_x").agg(avg_order=("total_amount","mean"),order_count=("order_id","count")).reset_index()
```
